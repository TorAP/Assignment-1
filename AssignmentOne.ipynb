```python
#Dividing the two lists 
ListLS = dl.loc[dl['Group']=="LS"]
ListL = dl.loc[dl['Group']=="L"]
```

```python
print('Group of people who were only listing to music')
ListL.describe()
```

```python
print('Group of people who were  listing to music AND seeing the visual')
ListLS.describe()
```

```python
#Another way to compare the two groups
group_by_audio = dl.groupby(by=['Group'])
ga =group_by_audio
ga_avg = group_by_audio.mean()
ga_median=group_by_audio.median()
ga_count = group_by_audio.count()

print('              MEAN values')
print(ga_avg)
print(80*'_')
print('              MEDIAN values')
print(ga_median)
print(80*'_')
print('              Number of data points')
print(ga_count)
```

```python
#The mean-values accros the five categories.

plt.plot(ListL.mean(), label ="Listening")
plt.plot(ListLS.mean(), label ='Listening and Seeing')

plt.ylabel('Percentage')
plt.title('Category')
plt.legend()
```

```python
#The median-values accros the five categories.
plt.plot(ListL.median(),label ="Listening")
plt.plot(ListLS.median(),label ='Listening and Seeing')
plt.ylabel('Percentage')
plt.title('Category')
plt.legend()
```

There is a difference between the two groups Listineing (L) and Listening and Seeing (LS). 

Looking at the median of the two groups, the group LS is generally more impressed of the 3 categories; skills, quality and pleasant.
